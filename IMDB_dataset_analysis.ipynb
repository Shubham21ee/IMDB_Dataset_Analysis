```python
#Classify movies based on the ratings(Execellent, Good and Average)
data.columns
```

```python
def rating(rating):
    if rating>=7.0:
        return "Excellent"
    elif rating >= 6.0:
        return "Good"
    else:
        return "Average" 
```

```python
data['raing_cat'] = data['Rating'].apply(rating)
```
